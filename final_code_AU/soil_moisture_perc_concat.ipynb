```python
# The percentiles are calculated separately for each day of year. Concatenate the files.

import xarray as xr
import numpy as np
import glob

main_dir = '/g/data/w97/ad9701/p_prob_analysis/processed_data/awra/'
list_perc = []
for day in range(1, 365+1):
    # if len(glob.glob(main_dir + 'sm_refPeriod_perc/by_doy/sm_*_perc_day' + str(day) + '.nc'))==0:
    #     print(day)
    perc_file = glob.glob(main_dir + 'sm_refPeriod_perc/by_doy/sm_*_perc_day' + str(day) + '.nc')[0]
    ds = xr.open_dataset(perc_file)
    list_perc.append(ds['sm'])
# Avoid using the 366th day as it is based on 1/4 the number of data points than all the other days. Substitute the 365th day for 366 instead.
if day == 365:
    list_perc.append(ds['sm'])
```

```python
out_file = main_dir + 'sm_refPeriod_perc/sm_percentiles.nc'

da_sm_perc = xr.concat(list_perc, dim='dayofyear')
dayofyear = np.arange(366)+1
da_sm_perc = da_sm_perc.assign_coords(dayofyear=dayofyear)
da_sm_perc.to_netcdf(out_file)
```
